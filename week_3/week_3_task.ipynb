```python
#Weekly_Task_3
#Analyze a Simple Sales Dataset to find: total sales, best-selling product, and create a basic report
import pandas as pd
# Sample sales data
data = pd.read_csv('train.csv')
# Create DataFrame
df = pd.DataFrame(data)
# Calculate total sales
total_sales = df['Sales'].sum()
# Find best-selling product
best_selling_product = df.groupby('Product Name')['Sales'].sum().idxmax()
best_selling_sales = df.groupby('Product Name')['Sales'].sum().max()
# Create report
report = f"Total Sales: ${total_sales}\nBest-Selling Product: {best_selling_product} with Sales: ${best_selling_sales}"
print(report)
# Save report to a text file
with open('sales_report.txt', 'w') as file:
    file.write(report)
```

```python
#Hands-on Practice
'''Install pandas library using simple pip command
Download a simple dataset (like student marks or sales data)
Load the dataset and display first few rows
Check basic information: how many rows, columns, data types
Find and handle missing values in the dataset
Calculate average, highest, lowest values for numerical columns'''
import pandas as pd
# Load dataset (using a sample dataset here)
data = pd.read_csv('iris-1.csv')  
# Display first few rows
print(data.head())
```

```python
# Check basic information
print(data.info())
```

```python
# Find and handle missing values
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Calculate average, highest, lowest values for numerical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    avg = data[col].mean()
    highest = data[col].max()
    lowest = data[col].min()
    print(f"{col} - Average: {avg}, Highest: {highest}, Lowest: {lowest}")
```
