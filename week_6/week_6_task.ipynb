Create an Interactive Sales Dashboard with multiple chart types showing sales trends, customer segmentation, and product performance

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('D:\Developer Arena/data.csv', encoding="latin1")
df
```

```python
df. describe()
```

```python
df.info()
```

```python
# Convert InvoiceDate to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
# Create TotalAmount column
df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
sns.set(style="whitegrid")
```

```python
#DISTRIBUTION OF QUANTITY
plt.figure(figsize=(8,5))
sns.histplot(df['Quantity'], bins=50, kde=True, color="#7b2cbf")
plt.title("Distribution of Quantity")
plt.show()
```

```python
#DISTRIBUTION OF UNIT PRICE

plt.figure(figsize=(8,5))
sns.histplot(df['UnitPrice'], bins=50, kde=True, color="#ff7f51")
plt.title("Distribution of Unit Price")
plt.show()
```

```python
#TOP 10 PRODUCTS BY SALES

top_products = df.groupby("Description")['Quantity'].sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette="magma")
plt.title("Top 20 Best Selling Products")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product")
plt.show()
```

```python
#UNIT PRICE vs QUANTITY
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df, x="UnitPrice", y="Quantity",
    alpha=0.3, color="#7209b7"
)
plt.title("Unit Price vs Quantity")
plt.show()
```

```python
#SALES COUNT PER COUNTRY (TOP 10)
country_count = df['Country'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(
    x=country_count.values,
    y=country_count.index,
    palette="Spectral"
)
plt.title("Top 10 Countries by Number of Sales Transactions")
plt.xlabel("Transaction Count")
plt.ylabel("Country")
plt.show()
```

```python
#REVENUE BY COUNTRY (TOP 10)

country_rev = (
    df.groupby("Country")['TotalAmount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))
sns.barplot(x=country_rev.values, y=country_rev.index, palette="coolwarm")
plt.title("Top 10 Revenue-Generating Countries")
plt.xlabel("Revenue")
plt.ylabel("Country")
plt.show()
```

```python
# REVENUE OVER TIME
daily_rev = df.groupby(df['InvoiceDate'].dt.date)['TotalAmount'].sum().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=daily_rev, x="InvoiceDate", y="TotalAmount", color="#38b000")
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

```python
#MONTHLY SALES TREND (Line Chart)

df["Month"] = df["InvoiceDate"].dt.to_period("M")
monthly_sales = df.groupby("Month")["TotalAmount"].sum()

plt.figure(figsize=(12,5))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# CORRELATION (ONLY NUMERICAL COLUMNS)
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['Quantity','UnitPrice','TotalAmount']].corr(),
    annot=True,
    cmap="Blues"
)
plt.title("Correlation Heatmap")
plt.show()
```
