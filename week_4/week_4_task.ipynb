```python
#Weekly Task 4: Weather Forecast Data Analysis and Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample student performance data
data = pd.read_csv('weather_forecast_data.csv')
data
```

```python
data.info()
```

1. Temprature Distribution: 
    The temperature distribution is centered around mid-range values, with no extreme spikes. Most temperature readings fall within 
a comfortable band, indicating stable weather conditions without frequent heatwaves or cold drops.

```python
# 1. Temperature Distribution
plt.figure()
plt.hist(data["Temperature"])
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

2. Humidity vs Temperature Scatter Plot:
    Humidity generally increases as temperature increases. This suggests a mild positive relationship, meaning warmer conditions often coincide with higher humidity, which is typical in moist or semi-tropical climates.

```python

# 2. Humidity vs Temperature Scatter
plt.figure()
plt.scatter(data["Temperature"], data["Humidity"])
plt.title("Humidity vs Temperature")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.tight_layout()
plt.show()
```

3. Wind Speed Distribution:
    Wind speed values are mostly low to moderate, with very few high-speed instances. This indicates calm weather patterns, with minimal strong or extreme wind events in the dataset.

```python
# 3. Wind Speed Distribution
plt.figure()
plt.hist(data["Wind_Speed"])
plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

4. Cloud Cover vs Pressure Scatter Plot:
    The plot shows no clear pattern or strong relationship between cloud cover and pressure. Pressure fluctuates independently, suggesting that cloudiness isnâ€™t directly influenced pressure variations in this dataset.

```python
# 4. Cloud Cover vs Pressure Scatter
plt.figure()
plt.scatter(data["Cloud_Cover"], data["Pressure"])
plt.title("Cloud Cover vs Pressure")
plt.xlabel("Cloud Cover")
plt.ylabel("Pressure")
plt.tight_layout()
plt.show()
```

5. Rain vs No Rain Bar Plot:
    The dataset shows more â€œno rainâ€ cases than â€œrain,â€ indicating mostly dry weather conditions. Rainfall events do occur but are significantly less frequent, reflecting generally stable and dry environmental patterns.

```python
# 5. Rain Count Bar Plot
rain_counts = data["Rain"].value_counts()

plt.figure()
plt.bar(rain_counts.index, rain_counts.values)
plt.title("Rain vs No Rain")
plt.xlabel("Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
